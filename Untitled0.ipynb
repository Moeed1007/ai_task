```python


class ShapeTrap:
    def __init__(self, p, q, h):
        self.p = p
        self.q = q
        self.h = h

    def get_area(self):
        return 0.5 * (self.p + self.q) * self.h


class ShapePara:
    def __init__(self, b, h):
        self.b = b
        self.h = h

    def get_area(self):
        return self.b * self.h


records = [
    (5, 8, 4, 6, 3),
    (7, 9, 5, 4, 6),
    (3, 6, 2, 8, 4)
]

step = 1

for row in records:
    a1, a2, a3, a4, a5 = row

    fig1 = ShapeTrap(a1, a2, a3)
    fig2 = ShapePara(a4, a5)

    save_area = {
        "Trapezoid": fig1.get_area(),
        "Parallelogram": fig2.get_area()
    }

    print("\nSet", step)
    print(save_area)

    top_one = max(save_area, key=save_area.get)

    print("Largest Shape:", top_one)
    print("Area:", save_area[top_one])

    step += 1
```
